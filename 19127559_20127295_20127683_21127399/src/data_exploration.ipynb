## Making your own question 01? (1.5 points)

**A possible question is :** How does the gender distribution vary by continent?


**Answering this question will:** show the differences in male and female populations across continents.

**How we answer this question:** Calculate the average male and female populations for each continent.

### Analyzing to get the answer

```python
year = 2022
data_for_fixed_year = df[df['Year'] == year]
# Group the data by continent and calculate the average male and female populations
average_gender_distribution = data_for_fixed_year.groupby('Country')[['Male Population', 'Female Population']].median().round(1)

# Display the result
print(average_gender_distribution)

plt.figure(figsize=(12, 10))
average_gender_distribution.plot(kind='line', stacked=True)
plt.title('Average Gender Distribution by Country in 2022')  # Đặt tiêu đề của biểu đồ
plt.xlabel('Country')  # Đặt nhãn trục x
plt.ylabel('Population')  # Đặt nhãn trục y
plt.legend(["Male Population", "Female Population"])  # Hiển thị chú thích
plt.show()
```

## Making your own question 04? (1.5 points)

**A possible question is :** What is top 5 country have the average percentage of students who complete primary school in each country?

**Answering this question will:** Show the top 5 averages of Primary completion rate for each continents.

**How we answer this question:** Calculate the average Primary completion rate for each continent.

```python
avg_Primary_completion_rate_df = pd.DataFrame()

# YOUR CODE HERE

# Find average for each country
avg_Primary_completion_rate_df["Avg.Primary completion rate"] = df.groupby("Country")["Primary completion rate"].mean()

# Set the index as a column - countries
avg_Primary_completion_rate_df["Country"] = avg_Primary_completion_rate_df.index

# Drop the index
avg_Primary_completion_rate_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Average Birth Rate
avg_Primary_completion_rate_df.sort_values("Avg.Primary completion rate", inplace=True, ascending=False)

# Display
avg_Primary_completion_rate_df.head()
```
